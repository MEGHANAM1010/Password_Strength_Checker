```python
#regular expressions or RegEx - built-in standard library used for adv. text processing , pattern matching , string manipulation
import re
def check_password_strength(password):
  issues=[]
  score=0
  #Here we are checking the length of the password
  if len(password)>=8:
    score+=1
  else:
    issues.append("Atleast 8 characters required")
  if len(password)>=12:
    score+=1
  #Here we are checking for different characters
  #Checking for digits
  if any(c.isdigit() for c in password):
    score+=1
  else:
    issues.append("Add atleast one number")
  #Checking for uppercase
  if any(c.isupper() for c in password):
    score+=1
  else:
    issues.append("Add atleast one uppercase character")
  #Checking for lowercase
  if any(c.islower() for c in password):
    score+=1
  else:
    issues.append("Add atleast one lowercase character")
  #checking for special characters
  #re.search() : Scans a string for the first match of a pattern anywhere in the text
  if re.search(r'[`~!@#$%^&*()_\-+={}\[\]|:<>,.?/]',password):
    score+=1
  else:
    issues.append("Add special characters to strengthen the password ")
  #Here we are checking the scores and rate each of them accordingly
  if score>=5: rating="🟢Strong Password"
  elif score>=3: rating="🟡Moderate Password"
  elif score>=1: rating="🟠Weak Password"
  else: rating="🔴Very Weak Password"
  #To print the result
  print("="*27)
  print("Rating:",rating)
  print("Score:",score,"/6")
  print("="*27)
  if issues:
    for i in issues:
      print(i)
  else:
      print("No issues found!")
  print("="*27)
#main() function
def password_checker():
  print("🔐Welcome to the Password Strength Checker!")
  while True:
    password=input("\n Enter your password (type 'exit' to quit):")
    if password.lower()=="exit":
      print("Goodbye!")
      break
    check_password_strength(password)
password_checker()
```
